ðŸ’¡ **Question 2**

Write a function that takes a list sorted in non-decreasing order and deletes any duplicate nodes from the list. The list should only be traversed once.

For example if the linked list is 11->11->11->21->43->43->60 then removeDuplicates() should convert the list to 11->21->43->60.

</aside>
Input:
LinkedList: 
11->11->11->21->43->43->60
Output:
11->21->43->60

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None


def removeDuplicates(head):
    current = head

    while current is not None and current.next is not None:
        if current.value == current.next.value:
            current.next = current.next.next
        else:
            current = current.next

    return head


# Function to print a linked list
def printLinkedList(head):
    current = head
    while current is not None:
        print(current.value, end="->")
        current = current.next
    print("None")


llist = Node(11)
llist.next = Node(11)
llist.next.next = Node(11)
llist.next.next.next = Node(21)
llist.next.next.next.next = Node(43)
llist.next.next.next.next.next = Node(43)
llist.next.next.next.next.next.next = Node(60)

print("Original linked list:")
printLinkedList(llist)


newLinkedList = removeDuplicates(llist)

print("Modified linked list:")
printLinkedList(newLinkedList)
```

Given a linked list of sizeÂ **N**. The task is to reverse everyÂ **k**Â nodes (where k is an input to the function) in the linked list. If the number of nodes is not a multiple ofÂ *k*Â then left-out nodes, in the end, should be considered as a group and must beÂ reversed (See Example 2 for clarification).

</aside>
Input:
LinkedList: 1->2->2->4->5->6->7->8
K = 4
Output:4Â 2Â 2Â 1Â 8Â 7Â 6Â 5
Explanation:
The first 4 elements 1,2,2,4 are reversed first
and then the next 4 elements 5,6,7,8. Hence, the
resultant linked list is 4->2->2->1->8->7->6->5.


```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None


def reverseKNodes(head, k):
    if k == 1 or head is None:
        return head

    prev = None
    current = head
    next_node = None
    count = 0

    # Reverse k nodes
    while current is not None and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    # Recursively reverse the remaining linked list
    if next_node is not None:
        head.next = reverseKNodes(next_node, k)

    return prev


# Function to print a linked list
def printLinkedList(head):
    current = head
    while current is not None:
        print(current.value, end="->")
        current = current.next
    print("None")

llist = Node(1)
llist.next = Node(2)
llist.next.next = Node(2)
llist.next.next.next = Node(4)
llist.next.next.next.next = Node(5)
llist.next.next.next.next.next = Node(6)
llist.next.next.next.next.next.next = Node(7)
llist.next.next.next.next.next.next.next = Node(8)

print("Original linked list:")
printLinkedList(llist)


k = 4
newLinkedList = reverseKNodes(llist, k)

print("Modified linked list:")
printLinkedList(newLinkedList)
```

Given a linked list, write a function to reverse every alternate k nodes (where k is an input to the function) in an efficient way. Give the complexity of your algorithm.

**Example:**

</aside>
Inputs:   1->2->3->4->5->6->7->8->9->NULL and k = 3
Output:   3->2->1->4->5->6->9->8->7->NULL.

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None


def reverseAlternateKNodes(head, k):
    if k == 1 or head is None:
        return head

    prev = None
    current = head
    next_node = None
    count = 0

    # Reverse alternate k nodes
    while current is not None and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    # Connect the reversed group to the next group
    if head is not None:
        head.next = current

    # Skip the next k nodes
    count = 0
    while current is not None and count < k - 1:
        current = current.next
        count += 1

    # Recursively reverse the next alternate k nodes
    if current is not None:
        current.next = reverseAlternateKNodes(current.next, k)

    return prev


# Function to print a linked list
def printLinkedList(head):
    current = head
    while current is not None:
        print(current.value, end="->")
        current = current.next
    print("None")


llist = Node(1)
llist.next = Node(2)
llist.next.next = Node(3)
llist.next.next.next = Node(4)
llist.next.next.next.next = Node(5)
llist.next.next.next.next.next = Node(6)
llist.next.next.next.next.next.next = Node(7)
llist.next.next.next.next.next.next.next = Node(8)
llist.next.next.next.next.next.next.next.next = Node(9)

print("Original linked list:")
printLinkedList(llist)

k = 3
newLinkedList = reverseAlternateKNodes(llist, k)

print("Modified linked list:")
printLinkedList(newLinkedList)
```

ðŸ’¡ **Question 5**

Given a linked list and a key to be deleted. Delete last occurrence of key from linked. The list may have duplicates.

**Examples**:

</aside>
Input:   1->2->3->5->2->10, key = 2
Output:  1->2->3->5->10

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None


def deleteLastOccurrence(head, key):
    if head is None:
        return head

    prev = None
    current = head
    last_occurrence = None
    last_occurrence_prev = None

    # Traverse the linked list to find the last occurrence of the key
    while current is not None:
        if current.value == key:
            last_occurrence = current
            last_occurrence_prev = prev
        prev = current
        current = current.next

    # If the key is found, remove the last occurrence
    if last_occurrence is not None:
        # If the last occurrence is the head of the linked list
        if last_occurrence_prev is None:
            head = last_occurrence.next
        else:
            last_occurrence_prev.next = last_occurrence.next

    return head


# Function to print a linked list
def printLinkedList(head):
    current = head
    while current is not None:
        print(current.value, end="->")
        current = current.next
    print("None")

llist = Node(1)
llist.next = Node(2)
llist.next.next = Node(3)
llist.next.next.next = Node(5)
llist.next.next.next.next = Node(2)
llist.next.next.next.next.next = Node(10)

print("Original linked list:")
printLinkedList(llist)

key = 2
newLinkedList = deleteLastOccurrence(llist, key)

print("Modified linked list:")
printLinkedList(newLinkedList)
```

Given two sorted linked lists consisting ofÂ **N**Â andÂ **M**Â nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.

**Examples:**

Input:Â a: 5->10->15,Â b: 2->3->20

Output:Â 2->3->5->10->15->20

Input:Â a: 1->1, b: 2->4

Output:Â 1->1->2->4

</aside>

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None


def mergeSortedLists(a, b):
    # Check if any of the lists is empty
    if a is None:
        return b
    if b is None:
        return a

    # Set the head of the merged list
    if a.value <= b.value:
        head = a
        current_a = a.next
        current_b = b
    else:
        head = b
        current_a = a
        current_b = b.next

    current_merged = head

    # Merge the lists
    while current_a is not None and current_b is not None:
        if current_a.value <= current_b.value:
            current_merged.next = current_a
            current_a = current_a.next
        else:
            current_merged.next = current_b
            current_b = current_b.next
        current_merged = current_merged.next

    # Attach the remaining nodes, if any
    if current_a is not None:
        current_merged.next = current_a
    if current_b is not None:
        current_merged.next = current_b

    return head


# Function to print a linked list
def printLinkedList(head):
    current = head
    while current is not None:
        print(current.value, end="->")
        current = current.next
    print("None")

a = Node(5)
a.next = Node(10)
a.next.next = Node(15)

b = Node(2)
b.next = Node(3)
b.next.next = Node(20)

print("Linked list a:")
printLinkedList(a)
print("Linked list b:")
printLinkedList(b)

mergedList = mergeSortedLists(a, b)

print("Merged list:")
printLinkedList(mergedList)
```

ðŸ’¡ **Question 7**

Given aÂ **Doubly Linked List**, the task is to reverse the given Doubly Linked List.

**Example:**

</aside>
Original Linked list 10 8 4 2
Reversed Linked list 2 4 8 10

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.prev = None
        self.next = None


def reverseDoublyLinkedList(head):
    # Check if the list is empty or has only one node
    if head is None or head.next is None:
        return head

    current = head
    prev = None

    # Swap next and prev pointers for each node
    while current is not None:
        next_node = current.next
        current.next = prev
        current.prev = next_node
        prev = current
        current = next_node

    # Update the head pointer to the new first node
    head = prev

    return head


# Function to print a doubly linked list
def printDoublyLinkedList(head):
    current = head
    while current is not None:
        print(current.value, end=" ")
        current = current.next
    print()

head = Node(10)
head.next = Node(8)
head.next.prev = head
head.next.next = Node(4)
head.next.next.prev = head.next
head.next.next.next = Node(2)
head.next.next.next.prev = head.next.next

print("Original doubly linked list:")
printDoublyLinkedList(head)

reversedHead = reverseDoublyLinkedList(head)

print("Reversed doubly linked list:")
printDoublyLinkedList(reversedHead)
```

Given a doubly linked list and a position. The task is to delete a node from given position inÂ a doubly linked list.

**Example 1:**

</aside>
Input:
LinkedList = 1 <--> 3 <--> 4
x = 3
Output:1 3
Explanation:After deleting the node at
position 3 (position starts from 1),
the linked list will be now as 1->3.


```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.prev = None
        self.next = None


def deleteNodeFromPosition(head, position):
    # Check if the list is empty
    if head is None:
        return None

    # Case 1: Deleting the first node
    if position == 1:
        head = head.next
        if head is not None:
            head.prev = None
        return head

    # Traverse the list to find the node at the given position
    current = head
    count = 1
    while current is not None and count < position:
        current = current.next
        count += 1

    # If position is greater than the length of the list
    if current is None:
        return head

    # Case 2: Deleting a node in the middle or at the end
    current.prev.next = current.next
    if current.next is not None:
        current.next.prev = current.prev

    return head


# Function to print a doubly linked list
def printDoublyLinkedList(head):
    current = head
    while current is not None:
        print(current.value, end=" ")
        current = current.next
    print()


head = Node(1)
node2 = Node(3)
node3 = Node(4)
head.next = node2
node2.prev = head
node2.next = node3
node3.prev = node2

print("Original doubly linked list:")
printDoublyLinkedList(head)

head = deleteNodeFromPosition(head, 3)

print("Doubly linked list after deleting node at position 3:")
printDoublyLinkedList(head)
```
